## Import des bibliothèques 

```python
import matplotlib.pyplot as plt 
import numpy as np 
import cv2 
```

## Import et pré traitement de l'image

**1-Import et normalisation**

```python
# Import de l'image en Niveaux de gris
image = cv2.imread("cameraman.tif",cv2.IMREAD_GRAYSCALE)

#Normalisation de l'image 
X1 = cv2.normalize(image,None,alpha=0,beta=1,norm_type=cv2.NORM_MINMAX)

#Enregistrer les images X1 normalisée sous le nom X1
path = 'X1.jpg'
cv2.imwrite(path,X1)

# Affichage de l'image en niveau de gris et l'image normalisée 
plt.subplot(1,2,1)
plt.imshow(image,cmap="gray")
plt.title("Image en Noir et blanc")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(X1,cmap="gray")
plt.title("Image en Noir et blanc normalisée")
plt.axis("off")
plt.show()
```

La normalisation sert à limiter la valeur des pixels entre alpha = 0 et beta = 1 

**2-Application du bruit blanc Gaussien de variance 0.001**

```python
#application du filtre gaussien avec une variance de 0.01 et un noyeau de taille 5px x 5px 
X2 = cv2.GaussianBlur(X1,(5,5),0.01)

#Enregistrer X2
path = 'X2.jpg'
cv2.imwrite(path,X2)

# Afficher l'image
plt.imshow(X2,cmap="gray")
plt.title("Image filtrée var=0.01")
plt.axis("off")
plt.show()
```

**3- Faire Varier la variance**

```python
#application du filtre gaussien avec une variance de 0.001 et un noyeau de taille 5px x 5px 
X2_0 = cv2.GaussianBlur(X1,(5,5),0)

#Affichage des images normalisée et  filtrée
plt.imshow(X2_0,cmap="gray")
plt.title("Image filtrée var=0")
plt.axis("off")
plt.show()
```

Aussi petite soit elle la variance à un role majeur dans la nettetée des contours car elle donne des indications à la fonction gaussienne sur la disperssion des pixels et permet de rendre les contours plus nets.

**Plus la variance est grande plus l'image est nette au nivrau des contours**


**4-Application d'un bruit poivre et sel appelée aussi medianblur()**

```python
#appliquer median Blur
X3 = cv2.medianBlur(X1,5)

#Enregistrer X3
path = 'X3.jpg'
cv2.imwrite(path,X3)

#Affichage des images normalisée et  filtrée
plt.imshow(X3,cmap="gray")
plt.title("Median Blur p=0.05")
plt.axis("off")
plt.show()
```

```python
#appliquer median Blur
X3_1= cv2.medianBlur(X1,1)

#Affichage des images normalisée et  filtrée
plt.imshow(X3_1,cmap="gray")
plt.title("Median Blur p=0.01")
plt.axis("off")
plt.show()
```

**Plus p est grand moins l'image est nette** 

**5-Afficher X1, X2, X3 et comparer**

```python
#Affichage et comparaison 

plt.subplot(1,3,1)
plt.imshow(X1,cmap='gray')
plt.title("X1")
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(X2,cmap='gray')
plt.title("X2")
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(X3,cmap='gray')
plt.title("X3")
plt.axis("off")
plt.show
```

Le filtra gaussien est celui qui donne le plus de details au contour le bruit median va rendre l'image plus floue à certains autres endroits
