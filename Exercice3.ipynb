## Import des bibliothèques 

```python
import matplotlib.pyplot as plt 
import numpy as np 
import cv2 
```

## Import et traitement des images 

```python
X1 = cv2.imread("X1.jpg",cv2.IMREAD_GRAYSCALE)
X2 = cv2.imread("X2.jpg",cv2.IMREAD_GRAYSCALE)
X3 = cv2.imread("X3.jpg",cv2.IMREAD_GRAYSCALE)
```

## Réponses 

**2-Application du filtre moyenneur sur X2**

```python
#Application du filtre moyenneur de taille (3,3)
Y2 = cv2.blur(X2,(3,3))

#Affichage de l'image
plt.imshow(Y2,cmap='gray')
plt.title("Filtre moyenneur")
plt.axis("off")
plt.show()
```

**3-Application du filtre median sur X3**

```python
Y3 = cv2.medianBlur(X3,3)

#Affichage de l'image
plt.imshow(Y3,cmap='gray')
plt.title("Filtre median")
plt.axis("off")
plt.show()
```

**Oui le bruit a été atténué** 

**4-PSNR**

```python
#Lire les image en tableaux de nombres 
psnr = cv2.PSNR(X1,X2)
psnr
```

Le psnr entre X1 et X2 est très haut 

```python
psnr = cv2.PSNR(X1,X3)
psnr
```

Le meilleur PSNR est celui entre l'image X1 et X3 car il est le plus bas mais reste mauvais car il sont tout les deux superieur à 20
